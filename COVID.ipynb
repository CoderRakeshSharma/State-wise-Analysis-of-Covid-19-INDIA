```python
import pandas as pd
import plotly.express as px
import dash
import dash_html_components as html 
import dash_core_components as dcc
import plotly.graph_objects as go
from dash.dependencies import Input, Output
global x
x=0


external_stylesheets = ['https://codepen.io/chriddyp/pen/bWLwgP.css']

app = dash.Dash(__name__, external_stylesheets=external_stylesheets)


colors = { 'background': 'lightgrey' }

data = pd.read_csv("covid19.csv")

astcc = data.sort_values(["Total Confirmed Cases"], ascending = False)
asac = data.sort_values(["Active Cases"], ascending = False)
asdc = data.sort_values(["Death"], ascending = False)
ascc = data.sort_values(["Cured/Discharged/Migrated"], ascending = False)

tstcc = data.sort_values(["Total Confirmed Cases"], ascending = False).head(7)
tsac = data.sort_values(["Active Cases"], ascending = False).head(7)
tsdc = data.sort_values(["Death"], ascending = False).head(7)
tscc = data.sort_values(["Cured/Discharged/Migrated"], ascending = False).head(7)

bstcc = data.sort_values(["Total Confirmed Cases"], ascending = True).head(7)
bsac = data.sort_values(["Active Cases"], ascending = True).head(7)
bsdc = data.sort_values(["Death"], ascending = True).head(7)
bscc = data.sort_values(["Cured/Discharged/Migrated"], ascending = True).head(7)



fig = go.Figure()

fig.update_yaxes(showgrid=False) 
fig.update_xaxes(showgrid=False)


app.layout=html.Div(style={'backgroundColor': colors['background']}, children=[html.Div(html.H1('COVID 19:INDIA',
                                                    style={'color':'white',           
                                                   'text-align':'center',
                                                   'background-color':'blue',
                                                   'border-style': 'double',
                                                   'display':'inline-block', 
                                                   'width':'100%'
                                                    })),
                   
                   
                   
                    html.Div(dcc.Dropdown(id='dropdown1',options=[{'label': 'ALL STATES and UT','value': 'as'},
                                                                  {'label': '7 STATES/UT WITH HIGH COVID CASES','value': 'ts'},
                                                                  {'label': '7 STATES/UT WITH LESS COVID CASES','value': 'bs'}],
                                                                  placeholder="CHOOSE AN OPTION"),
                                                                     
                                                                  style={'width':'30%','display':'inline-block'}),
                    html.Div(dcc.Dropdown(id='dropdown2',options=[{'label': 'TOTAL CONFIRMED CASES ','value': 'tcc'},
                                                                  {'label': 'ACTIVE CASES','value': 'ac'},
                                                                  {'label': 'DEATH CASES ','value': 'dc'},
                                                                  {'label': 'RECOVERED CASES','value': 'cc'}],
                                                                  placeholder="CHOOSE AN OPTION"),                                                                  
                                                                  style={'width':'30%','display':'inline-block'}),
                    html.Div(dcc.Dropdown(id='dropdown3',options=[{'label': 'MAP','value': 'mp'},                                                                  
                                                                  {'label': 'GRAPH','value': 'gr'}],
                                                                  placeholder="CHOOSE AN OPTION"),                                                                  
                                                                  style={'width':'30%','display':'inline-block'}),
                    html.Div(html.Button('CONFIRM', id='button',n_clicks=0,style={'width':'10%','display':'inline-block','color':'hotpink','background-color':'blue'})),
               
                    html.Div(dcc.Graph(id='firstgraph1',figure=fig,style={'width':'100%','display':'inline-block'})),
                    html.Div(html.H1('CREATED BY RAKESH SHARMA',
                                                    style={'color':'white',           
                                                   'text-align':'center',
                                                   'background-color':'blue',
                                                   'border-style': 'double',
                                                   'display':'inline-block', 
                                                   'width':'100%'
                                                    }))
                    ])
                     
@app.callback(
    Output('firstgraph1', 'figure'),
    [Input('dropdown1', 'value'),Input('dropdown2', 'value'),Input('dropdown3', 'value'),Input('button', 'n_clicks'),]) 

def update_figure(dropdown1,dropdown2,dropdown3,n_clicks):
    global x
    global fig
    if (dropdown1=='as') & (dropdown2=='tcc') & (dropdown3=='gr') & (n_clicks>x) :
        fig= go.Figure(data=go.Scatter(x=astcc['Name Of State or UT'],y=astcc['Total Confirmed Cases'],mode="lines + markers",marker=dict(color='black')))
        fig.update_layout(title="TOTAL CONFIRMED  CASES IN ALL STATES AND UT",xaxis_title="State or UT",yaxis_title="Total Confirmed Cases")        
    if (dropdown1=='as') & (dropdown2=='ac') & (dropdown3=='gr') & (n_clicks>x) :
        fig= go.Figure(data=go.Scatter(x=asac['Name Of State or UT'],y=asac['Active Cases'],mode="lines + markers",marker=dict(color='navy')))
        fig.update_layout(title="ACTIVE CASES IN ALL STATES AND UT",xaxis_title="State or UT",yaxis_title="Active Cases")
        
    if (dropdown1=='as') & (dropdown2=='dc') & (dropdown3=='gr') & (n_clicks>x) :
        fig= go.Figure(data=go.Scatter(x=asdc['Name Of State or UT'],y=asdc['Death'],mode="lines + markers",marker=dict(color='red')))
        fig.update_layout(title="DEATH CASES IN ALL STATES AND UT",xaxis_title="State or UT",yaxis_title="Death Cases")
        
    if (dropdown1=='as') & (dropdown2=='cc') & (dropdown3=='gr') & (n_clicks>x) :
        fig= go.Figure(data=go.Scatter(x=ascc['Name Of State or UT'],y=ascc['Cured/Discharged/Migrated'],mode="lines + markers",marker=dict(color='green')))
        fig.update_layout(title="RECOVERED CASES IN ALL STATES AND UT",xaxis_title="State or UT",yaxis_title="Recovered Cases")
        
   
        
    if (dropdown1=='ts') & (dropdown2=='tcc') & (dropdown3=='gr') & (n_clicks>x) :
        fig= go.Figure(data=go.Bar(x=tstcc['Name Of State or UT'],y=tstcc['Total Confirmed Cases'],
                                   text = tstcc['Total Confirmed Cases'],textposition = 'auto'))
        fig.update_traces(marker_color = 'dodgerblue')
        fig.update_layout(title="7 STATES OR UT WITH HIGHEST TOTAL CONFIRMED  CASES",xaxis_title="State or UT",yaxis_title="Total Confirmed Cases")
        
    if (dropdown1=='ts') & (dropdown2=='ac') & (dropdown3=='gr') & (n_clicks>x) :
        fig= go.Figure(data=go.Bar(x=tsac['Name Of State or UT'],y=tsac['Active Cases'],
                                  text = tsac['Active Cases'],textposition = 'auto'))
        fig.update_traces(marker_color = 'darkolivegreen')
        fig.update_layout(title="7 STATES OR UT WITH HIGHEST ACTIVE CASES",xaxis_title="State or UT",yaxis_title="Active Cases")
        
    if (dropdown1=='ts') & (dropdown2=='dc') & (dropdown3=='gr') & (n_clicks>x) :
        fig= go.Figure(data=go.Bar(x=tsdc['Name Of State or UT'],y=tsdc['Death'],
                                  text = tsdc['Death'],textposition = 'auto'))
        fig.update_traces(marker_color = 'purple')
        fig.update_layout(title="7 STATES OR UT WITH HIGHEST DEATH CASES",xaxis_title="State or UT",yaxis_title="Death Cases")
        
    if (dropdown1=='ts') & (dropdown2=='cc') & (dropdown3=='gr') & (n_clicks>x) :
        fig= go.Figure(data=go.Bar(x=tscc['Name Of State or UT'],y=tscc['Cured/Discharged/Migrated'],
                                  text = tscc['Cured/Discharged/Migrated'],textposition = 'auto'))
        fig.update_traces(marker_color = 'maroon')
        fig.update_layout(title="7 STATES OR UT WITH HIGH RECOVERY CASES",xaxis_title="State or UT",yaxis_title="Recovered Cases")
        
   
    
    if (dropdown1=='bs') & (dropdown2=='tcc') & (dropdown3=='gr') & (n_clicks>x) :
        fig= go.Figure(data=go.Bar(x=bstcc['Name Of State or UT'],y=bstcc['Total Confirmed Cases'],
                                  text = bstcc['Total Confirmed Cases'],textposition = 'auto'))
        fig.update_traces(marker_color = 'forestgreen')
        fig.update_layout(title="7 STATES OR UT WITH LEAST TOTAL CONFIRMED  CASES",xaxis_title="State or UT",yaxis_title="TOTAL CONFIRMED  CASES")
        
    if (dropdown1=='bs') & (dropdown2=='ac') & (dropdown3=='gr') & (n_clicks>x) :
        fig= go.Figure(data=go.Bar(x=bsac['Name Of State or UT'],y=bsac['Active Cases'],
                                  text = bsac['Active Cases'],textposition = 'auto'))
        fig.update_traces(marker_color = 'navy')
        fig.update_layout(title="7 STATES OR UT WITH LEAST ACTIVE CASES",xaxis_title="State or UT",yaxis_title="Active Cases")
        
    if (dropdown1=='bs') & (dropdown2=='dc') & (dropdown3=='gr') & (n_clicks>x) :
        fig= go.Figure(data=go.Bar(x=bsdc['Name Of State or UT'],y=bsdc['Death'],
                                  text = bsdc['Death'],textposition = 'auto'))
        fig.update_traces(marker_color = 'royalblue')
        fig.update_layout(title="7 STATES OR UT WITH LEAST DEATH CASES",xaxis_title="State or UT",yaxis_title="Death Cases")
        
    if (dropdown1=='bs') & (dropdown2=='cc') & (dropdown3=='gr') & (n_clicks>x) :
        fig= go.Figure(data=go.Bar(x=bscc['Name Of State or UT'],y=bscc['Cured/Discharged/Migrated'],
                                  text = bscc['Cured/Discharged/Migrated'],textposition = 'auto'))
        fig.update_traces(marker_color = 'orangered')
        fig.update_layout(title="7 STATES OR UT WITH LESS RECOVERY CASES",xaxis_title="State or UT",yaxis_title="Recovered Cases")
        
        
        
    if (dropdown1=='as') & (dropdown2=='tcc') & (dropdown3=='mp') & (n_clicks>x) :
        fig = px.choropleth(
             astcc,
             geojson="https://gist.githubusercontent.com/jbrobst/56c13bbbf9d97d187fea01ca62ea5112/raw/e388c4cae20aa53cb5090210a42ebb9b765c0a36/india_states.geojson",
             featureidkey='properties.ST_NM',
             locations='Name Of State or UT',
             color='Total Confirmed Cases',
   
             color_continuous_scale='oranges'
            )
        fig.update_geos(fitbounds="locations", visible=True)
    
    if (dropdown1=='as') & (dropdown2=='ac') & (dropdown3=='mp') & (n_clicks>x) :
        fig = px.choropleth(
             asac,
             geojson="https://gist.githubusercontent.com/jbrobst/56c13bbbf9d97d187fea01ca62ea5112/raw/e388c4cae20aa53cb5090210a42ebb9b765c0a36/india_states.geojson",
             featureidkey='properties.ST_NM',
             locations='Name Of State or UT',
             color='Active Cases',
   
             color_continuous_scale='magenta'
            )
        fig.update_geos(fitbounds="locations", visible=True)
        
    if (dropdown1=='as') & (dropdown2=='dc') & (dropdown3=='mp') & (n_clicks>x) :
        fig = px.choropleth(
             asdc,
             geojson="https://gist.githubusercontent.com/jbrobst/56c13bbbf9d97d187fea01ca62ea5112/raw/e388c4cae20aa53cb5090210a42ebb9b765c0a36/india_states.geojson",
             featureidkey='properties.ST_NM',
             locations='Name Of State or UT',
             color='Death',
   
             color_continuous_scale='reds'
            )
        fig.update_geos(fitbounds="locations", visible=True)
        
    if (dropdown1=='as') & (dropdown2=='cc') & (dropdown3=='mp') & (n_clicks>x) :
        fig = px.choropleth(
             ascc,
             geojson="https://gist.githubusercontent.com/jbrobst/56c13bbbf9d97d187fea01ca62ea5112/raw/e388c4cae20aa53cb5090210a42ebb9b765c0a36/india_states.geojson",
             featureidkey='properties.ST_NM',
             locations='Name Of State or UT',
             color='Cured/Discharged/Migrated',
   
             color_continuous_scale='greens'
            )
        fig.update_geos(fitbounds="locations", visible=True)
        
        
    if (dropdown1=='ts') & (dropdown2=='tcc') & (dropdown3=='mp') & (n_clicks>x) :
        fig = px.choropleth(
             tstcc,
             geojson="https://gist.githubusercontent.com/jbrobst/56c13bbbf9d97d187fea01ca62ea5112/raw/e388c4cae20aa53cb5090210a42ebb9b765c0a36/india_states.geojson",
             featureidkey='properties.ST_NM',
             locations='Name Of State or UT',
             color='Total Confirmed Cases',
   
             color_continuous_scale='tropic'
            )
        fig.update_geos(fitbounds="locations", visible=True)
    
    if (dropdown1=='ts') & (dropdown2=='ac') & (dropdown3=='mp') & (n_clicks>x) :
        fig = px.choropleth(
             tsac,
             geojson="https://gist.githubusercontent.com/jbrobst/56c13bbbf9d97d187fea01ca62ea5112/raw/e388c4cae20aa53cb5090210a42ebb9b765c0a36/india_states.geojson",
             featureidkey='properties.ST_NM',
             locations='Name Of State or UT',
             color='Active Cases',
   
             color_continuous_scale='balance'
            )
        fig.update_geos(fitbounds="locations", visible=True)
        
    if (dropdown1=='ts') & (dropdown2=='dc') & (dropdown3=='mp') & (n_clicks>x) :
        fig = px.choropleth(
             tsdc,
             geojson="https://gist.githubusercontent.com/jbrobst/56c13bbbf9d97d187fea01ca62ea5112/raw/e388c4cae20aa53cb5090210a42ebb9b765c0a36/india_states.geojson",
             featureidkey='properties.ST_NM',
             locations='Name Of State or UT',
             color='Death',
   
             color_continuous_scale='brbg'
            )
        fig.update_geos(fitbounds="locations", visible=True)
        
    if (dropdown1=='ts') & (dropdown2=='cc') & (dropdown3=='mp') & (n_clicks>x) :
        fig = px.choropleth(
             tscc,
             geojson="https://gist.githubusercontent.com/jbrobst/56c13bbbf9d97d187fea01ca62ea5112/raw/e388c4cae20aa53cb5090210a42ebb9b765c0a36/india_states.geojson",
             featureidkey='properties.ST_NM',
             locations='Name Of State or UT',
             color='Cured/Discharged/Migrated',
   
             color_continuous_scale='edge'
            )
        fig.update_geos(fitbounds="locations", visible=True)
        
        
    if (dropdown1=='bs') & (dropdown2=='tcc') & (dropdown3=='mp') & (n_clicks>x) :
        fig = px.choropleth(
             bstcc,
             geojson="https://gist.githubusercontent.com/jbrobst/56c13bbbf9d97d187fea01ca62ea5112/raw/e388c4cae20aa53cb5090210a42ebb9b765c0a36/india_states.geojson",
             featureidkey='properties.ST_NM',
             locations='Name Of State or UT',
             color='Total Confirmed Cases',
   
             color_continuous_scale='thermal'
            )
        fig.update_geos(fitbounds="locations", visible=True)
    
    if (dropdown1=='bs') & (dropdown2=='ac') & (dropdown3=='mp') & (n_clicks>x) :
        fig = px.choropleth(
             bsac,
             geojson="https://gist.githubusercontent.com/jbrobst/56c13bbbf9d97d187fea01ca62ea5112/raw/e388c4cae20aa53cb5090210a42ebb9b765c0a36/india_states.geojson",
             featureidkey='properties.ST_NM',
             locations='Name Of State or UT',
             color='Active Cases',
   
             color_continuous_scale='purpor'
            )
        fig.update_geos(fitbounds="locations", visible=True)
        
    if (dropdown1=='bs') & (dropdown2=='dc') & (dropdown3=='mp') & (n_clicks>x) :
        fig = px.choropleth(
             bsdc,
             geojson="https://gist.githubusercontent.com/jbrobst/56c13bbbf9d97d187fea01ca62ea5112/raw/e388c4cae20aa53cb5090210a42ebb9b765c0a36/india_states.geojson",
             featureidkey='properties.ST_NM',
             locations='Name Of State or UT',
             color='Death',
   
             color_continuous_scale='earth'
            )
        fig.update_geos(fitbounds="locations", visible=True)
        
    if (dropdown1=='bs') & (dropdown2=='cc') & (dropdown3=='mp') & (n_clicks>x) :
        fig = px.choropleth(
             bscc,
             geojson="https://gist.githubusercontent.com/jbrobst/56c13bbbf9d97d187fea01ca62ea5112/raw/e388c4cae20aa53cb5090210a42ebb9b765c0a36/india_states.geojson",
             featureidkey='properties.ST_NM',
             locations='Name Of State or UT',
             color='Cured/Discharged/Migrated',
   
             color_continuous_scale='icefire'
            )
        fig.update_geos(fitbounds="locations", visible=True)
   
    
    
    x=n_clicks
    fig.layout.plot_bgcolor = 'lightgrey'
    fig.layout.paper_bgcolor = 'lightgrey'
    return fig


if __name__ == '__main__':        
    app.run_server()
```
